# Multi-Head Latent Attention (MLA) Cache Analyzer
This notebook visualizes the KV cache memory math and proves when MLA is mathematically mandatory to avoid OOM.

```python
import numpy as np

def calculate_kv_cache_size(batch_size, seq_len, num_layers, num_heads, head_dim, precision_bytes=2):
    """
    Calculate memory required for MHA KV cache.
    Formula: 2 (K & V) * batch_size * seq_len * num_layers * num_heads * head_dim * precision_bytes
    """
    return 2 * batch_size * seq_len * num_layers * num_heads * head_dim * precision_bytes

# Example: Llama-3 70B (approx dims)
mha_size = calculate_kv_cache_size(batch_size=128, seq_len=8192, num_layers=80, num_heads=8, head_dim=128)
print(f"MHA KV Cache Size: {mha_size / 1e9:.2f} GB")
```
